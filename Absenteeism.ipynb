# Removing Outliers

```python
from scipy.stats import zscore
z_score=zscore(df)
print(df.shape)
df_1=df.loc[(z_score<3).all(axis=1)]
print(df_1.shape)
```

# Skewness

```python
df_1.skew()
```

```python
plt.figure(figsize=(10,8))
plot=1
for i in df.columns:
    if plot<=25:
        plt.subplot(5,5,plot)
        sns.distplot(df[i])
    plot=plot+1
plt.show()    
```

```python
for col in df_1.columns:
    if df_1[col].skew()>.55:
        df_1[col]=np.log1p(df[col])
    if df_1[col].skew()<-.55:
        df_1[col]=np.log1p(df[col])
```

```python
df_1.skew()
```

```python
## Dropping the Column
df_1.drop('ID',axis=1,inplace=True)
```

# Assigning X And Y Variable  

```python
x=df_1.drop('Absenteeism time in hours',axis=1)
y=df_1[['Absenteeism time in hours']]
```

```python
## Checking Mutlicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x_scaled=ss.fit_transform(x)
x_scaled=pd.DataFrame(x_scaled,columns=x.columns)
x_scaled
```

```python
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_scaled,i) for i in range (x_scaled.shape[1])]
vif['Feature']=x.columns

vif
```

```python
x_scaled.drop(['Weight','Height'],axis=1,inplace=True)
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error,r2_score
```

```python
def maxr2_score(clf,x,y):
    max_r=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=r_state)
        clf.fit(x_train,y_train)
        pred=clf.predict(x_test)
        r2_scr=r2_score(y_test,pred)
        print('r2_score corresponding to random state:',r_state,'is:',r2_scr)
        if r2_scr>max_r:
            max_r=r2_scr
            final_r_state=r_state
    print()
    print()
    print('max r2_score correspondint to :',final_r_state,'is',max_r)
    return final_r_state
```

```python
### Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
r_state=maxr2_score(lr,x_scaled,y)
```

```python
from sklearn.linear_model import Ridge,Lasso
rig=Ridge()
lasso=Lasso()
```

```python
#Ridge
r_state=maxr2_score(rig,x_scaled,y)
```

# KNN

```python
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
param={'n_neighbors':[11,21,25]}
grid=GridSearchCV(knn,param,scoring='r2',cv=7)
grid.fit(x_scaled,y)
grid.best_params_
```

```python
knn=KNeighborsRegressor(n_neighbors=11)
r_state=maxr2_score(knn,x_scaled,y)
```

# Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
rff=RandomForestRegressor()
param={'n_estimators':[10,50,100,200,500],'min_samples_leaf':[2,4,8]}
gdd=GridSearchCV(rff,param,cv=7,scoring='r2')
gdd.fit(x_scaled,y)
gdd.best_params_
```

```python
rff=RandomForestRegressor(min_samples_leaf= 8, n_estimators= 100)
maxr2_score(rff,x_scaled,y)
```

```python
### SVM
from sklearn.svm import SVR
svr=SVR()
param={'kernel':['poly','linear','rbf'],'C':[.0001,.001,.01,1,10]}
gdd=GridSearchCV(svr,param,scoring='r2',cv=7)
gdd.fit(x_scaled,y)
gdd.best_params_
```

```python
svr=SVR(C= 0.01,kernel='linear')
maxr2_score(svr,x_scaled,y)
```

```python
### Gradient Boosting Regressor
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
parameters = {'learning_rate': [0.001,0.01,0.1,0.02,0.03,1],
              'subsample'    : [0.9, 0.5, 0.2],
              'n_estimators' : [100,500,1000],
              'max_depth'    : [4,6,8]}
grid = GridSearchCV(estimator=gbr, param_grid = parameters, cv = 7, n_jobs=-1)
grid.fit(x_scaled, y)
grid.best_params_
```

```python
gbr=GradientBoostingRegressor(learning_rate=0.01, max_depth= 4, n_estimators= 500,subsample= 0.9)
maxr2_score(gbr,x_scaled,y)
```

# Cross Validatio Score

```python
##Gradiant Boosting
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,random_state=93,test_size=.20)
gbr.fit(x_train,y_train)
y_pred=gbr.predict(x_test)
print('Cross Validation score:',cross_val_score(gbr,x_scaled,y,scoring='r2',cv=7))
print('Mean cross Val Score:',cross_val_score(gbr,x_scaled,y,scoring='r2',cv=7).mean())
```

```python
## Random Forest
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,random_state=77,test_size=.20)
rff.fit(x_train,y_train)
y_pred=rff.predict(x_test)
print('Cross Validation score:',cross_val_score(rff,x_scaled,y,scoring='r2',cv=7))
print('Mean cross Val Score:',cross_val_score(rff,x_scaled,y,scoring='r2',cv=7).mean())
```

```python
## Error
print('MAE of Gradiant Boost Regressor:',mean_absolute_error(y_test,y_pred))
print('MSE of Gradiant Boost Regressor:',mean_squared_error(y_test,y_pred))
print('Roor suared Error:',np.sqrt(mean_squared_error(y_test,y_pred)))
```
