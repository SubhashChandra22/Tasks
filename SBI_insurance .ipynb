```python
## Lets seperate the target variable and input data
x=df.drop('insuranceclaim',axis=1)
y=df[['insuranceclaim']]
```

```python
##Lets rescaling the features to common scale as we will use logistic,knn,svm algo
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_scaled=ss.fit_transform(x)
X_scaled=pd.DataFrame(X_scaled,columns=x.columns)
```

```python
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.model_selection import train_test_split
def maxf1_score(clf,X_scaled,y):
    maxf=0
    rs=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(X_scaled,y,test_size=.20,random_state=r_state,stratify=y)
        clf.fit(x_train,y_train)
        y_pred=clf.predict(x_test)
        tep=f1_score(y_test,y_pred)
        if tep>maxf:
            maxf=tep
            rs=r_state
    print('Maximun f1_score is at random state:',rs,'and it is :',maxf)
    print('Confusion_matrix:',confusion_matrix(y_test,y_pred))
```

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression
lg_clf=LogisticRegression()
maxf1_score(lg_clf,X_scaled,y)
```

```python
## cross validation
from sklearn.model_selection import cross_val_score
def cross_val(clf,X_scaled,y):
    print('Mean cross validation f1_score:',cross_val_score(clf,X_scaled,y,cv=5,scoring='f1').mean())
    print('Standard Deviation:',cross_val_score(clf,X_scaled,y,cv=5,scoring='f1').std())
    print('Cross Val score:',cross_val_score(clf,X_scaled,y,cv=5,scoring='f1'))
```

```python
cross_val(lg_clf,X_scaled,y)
```

```python
##Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt_clf=DecisionTreeClassifier()
maxf1_score(dt_clf,X_scaled,y)
```

```python
## Cross validation
cross_val(dt_clf,X_scaled,y)
```

```python
### Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
parameters={'n_estimators':[10,100,500],'criterion':['gini','entorpy']}
rf_clf=RandomForestClassifier()
clf=GridSearchCV(estimator=rf_clf,param_grid=parameters,cv=5,scoring='f1')
clf.fit(X_scaled,y)
clf.best_params_
```

```python
##random Forest
rf_clf=RandomForestClassifier(n_estimators=500,criterion='gini')
maxf1_score(rf_clf,X_scaled,y)
```

```python
## Cross validation of Random Forest
cross_val(rf_clf,X_scaled,y)
```

```python
## Decision Tree classifier works well as compaire to other models
from sklearn.metrics import classification_report,roc_auc_score
x_train,x_test,y_train,y_test=train_test_split(X_scaled,y,test_size=.20,random_state=500,stratify=y)
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
print('f_1 score:',f1_score(y_test,y_pred))
print('Confusion Matrix \n:',confusion_matrix(y_test,y_pred))
print('Classification report \n',classification_report(y_test,y_pred))
print('AUC-ROC score:',roc_auc_score(y_pred,y_test))
```
